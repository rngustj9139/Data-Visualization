```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
fig = plt.figure(figsize=(7, 7), facecolor='linen')

ax1 = plt.subplot2grid((1, 1), (0, 0), fig=fig) # 행렬(행1개, 열1개) 만들고 지금 내 위치는 (0, 0)이라는 것을 의미
```

```python
fig = plt.figure(figsize=(7, 7), facecolor='linen')

ax1 = plt.subplot2grid((2, 1), (0, 0), fig=fig)
ax2 = plt.subplot2grid((2, 1), (1, 0), fig=fig)
```

```python
fig = plt.figure(figsize=(7, 7), facecolor='linen')

ax1 = plt.subplot2grid((3, 3), (0, 0), colspan=2, fig=fig)
# ax2 = plt.subplot2grid((3, 3), (0, 1), fig=fig)
ax3 = plt.subplot2grid((3, 3), (0, 2), fig=fig)
ax4 = plt.subplot2grid((3, 3), (1, 0), fig=fig)
ax5 = plt.subplot2grid((3, 3), (1, 1), fig=fig)
ax6 = plt.subplot2grid((3, 3), (1, 2), fig=fig)
ax7 = plt.subplot2grid((3, 3), (2, 0), fig=fig)
ax8 = plt.subplot2grid((3, 3), (2, 1), fig=fig)
ax9 = plt.subplot2grid((3, 3), (2, 2), fig=fig)
```

```python
fig = plt.figure(figsize=(7, 7), facecolor='linen')

ax1 = plt.subplot2grid((3, 3), (0, 0), colspan=2, fig=fig)
ax2 = plt.subplot2grid((3, 3), (1, 0), colspan=2, fig=fig)
ax3 = plt.subplot2grid((3, 3), (2, 0), colspan=2, fig=fig)
ax4 = plt.subplot2grid((3, 3), (0, 2), rowspan=3, fig=fig)
```
