```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
fig = plt.figure(figsize=(7, 7), facecolor='linen')

ax1 = fig.add_subplot(3, 1, 1)
ax2 = fig.add_subplot(3, 1, 2)
ax3 = fig.add_subplot(3, 1, 3)
```

```python
fig, axes = plt.subplots(nrows=3, ncols=1, figsize=(7, 7), facecolor='linen')
```

```python
fig = plt.figure(figsize=(7, 7), facecolor='linen')

ax1 = plt.subplot2grid((3, 1), (0, 0), fig=fig)
ax2 = plt.subplot2grid((3, 1), (1, 0), fig=fig)
ax3 = plt.subplot2grid((3, 1), (2, 0), fig=fig)
```

```python
fig = plt.figure(figsize=(7, 7), facecolor='linen')

ax1 = fig.add_subplot(1, 2, 1)
ax2 = fig.add_subplot(2, 2, 2)
ax3 = fig.add_subplot(2, 2, 4)
```

```python
fig = plt.figure(figsize=(7, 7), facecolor='linen')

ax1 = plt.subplot2grid((2, 2), (0, 0), rowspan=2, fig=fig)
ax2 = plt.subplot2grid((2, 2), (0, 1), fig=fig)
ax3 = plt.subplot2grid((2, 2), (1, 1), fig=fig)
```

```python
fig = plt.figure(figsize=(7, 7), facecolor='linen')

axes = np.empty(shape=(0,))
axes = np.append(axes, plt.subplot2grid((5, 4), (0, 0), rowspan=2, colspan=2, fig=fig))
axes = np.append(axes, plt.subplot2grid((5, 4), (0, 2), rowspan=2, colspan=2, fig=fig))

for r_idx in range(2, 5):
    for c_idx in range(4):
        axes = np.append(axes, plt.subplot2grid((5, 4), (r_idx, c_idx), fig=fig))
```
