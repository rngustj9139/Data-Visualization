```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
fig, axes = plt.subplots(nrows=1, ncols=1, figsize=(7, 7), facecolor='linen')

axes.set_title('Title!', fontsize=20)
axes.set_xlabel('X label!', fontsize=15)
axes.set_ylabel('Y label!', fontsize=15)
```

```python
title_list = ['Title' + str(i) for i in range(4)]
xlabel_list = ['X label' + str(i) for i in range(4)]
ylabel_list = ['Y label' + str(i) for i in range(4)]

fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(7, 7), facecolor='linen')

for ax_idx, ax in enumerate(axes.flat):
    ax.set_title(title_list[ax_idx], fontsize=15)
    ax.set_xlabel(xlabel_list[ax_idx], fontsize=15)
    ax.set_ylabel(ylabel_list[ax_idx], fontsize=15)
```

```python
title_list = ['Title' + str(i) for i in range(4)]
xlabel_list = ['X label' + str(i) for i in range(4)]
ylabel_list = ['Y label' + str(i) for i in range(4)]

fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(7, 7), facecolor='linen')

# fig.tight_layout()
fig.tight_layout(pad=7)

for ax_idx, ax in enumerate(axes.flat):
    ax.set_title(title_list[ax_idx], fontsize=15)
    ax.set_xlabel(xlabel_list[ax_idx], fontsize=15)
    ax.set_ylabel(ylabel_list[ax_idx], fontsize=15)
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(7, 7), facecolor='linen')

for ax_idx, ax in enumerate(axes.flat):
    xaxis = ax.get_xaxis()
    yaxis = ax.get_yaxis()
    
    xaxis.set_visible(False)
    yaxis.set_visible(False)
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(7, 7), facecolor='linen')

for ax_idx, ax in enumerate(axes.flat):
    ax.get_xaxis().set_visible(False)
    ax.get_yaxis().set_visible(False)
    
fig.subplots_adjust(bottom=0.01, top=0.99) #  0.01  top 0.99  
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(7, 7), facecolor='linen')

for ax_idx, ax in enumerate(axes.flat):
    ax.get_xaxis().set_visible(False)
    ax.get_yaxis().set_visible(False)
    
fig.subplots_adjust(left=0.01, right=0.99) #  0.01  top 0.99  
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(7, 7), facecolor='linen')

for ax_idx, ax in enumerate(axes.flat):
    ax.get_xaxis().set_visible(False)
    ax.get_yaxis().set_visible(False)
    
fig.subplots_adjust(bottom=0.01, top=0.99, left=0.01, right=0.99) #  0.01  top 0.99  
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(7, 7), facecolor='linen')

for ax_idx, ax in enumerate(axes.flat):
    ax.get_xaxis().set_visible(False)
    ax.get_yaxis().set_visible(False)
    
fig.subplots_adjust(hspace=0.05, wspace=0.05)
```
